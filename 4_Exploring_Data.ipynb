Exploring Data
===
When you get a dataset for the first time you'll want to do a few things to characterize it. This initial exploration sometimes generates interesting insights and may help you decide where to start with your analysis and what approaches are best suited or which are not likely to work well. Most of these exploratory techniques are just ways to visualize the dataset, but we'll also touch on ways to quantify what we're looking at

Histogram, Box Plots, and Distributions
---
The first thing you'll often want to do with a new dataset is to know how it's distributed. Specifically, you might be interested in whether or not that distribution is normal. A good way to visualize the distribution of a dataset is with a histogram. 

**Exercise 5.1:** Below we create an array of 1000 normally distributed values and plot them as a histogram. What does the histogram look like when there are fewer values (say 50)? What happens if you increase or decrease the number of histogram bins using the `bin` parameter of the `hist` method?

```python
x_norm = np.random.randn(1000)  # create an array of 1000 normally distributed values
pyplot.hist(x_norm)
pyplot.title('normal distribution')
pyplot.xlabel("values")
pyplot.show()
```

Histograms are best suited for looking at the distribution one or two one-dimensional data sets at a time ([here's](http://stackoverflow.com/questions/6871201/plot-two-histograms-at-the-same-time-with-matplotlib) an example of how to plot two histograms at once). Use a boxplot to compare 4 or more distributions in one window.

```python
# create an array of 1000 normally distributed values, mean of 1, standard deviation of 1
w_norm = 1 * np.random.randn(1000) + 1

# create an array of 1000 normally distributed values, mean of 10, standard deviation of 3
x_norm = 3 * np.random.randn(1000) + 10

# create an array of 1000 normally distributed values, mean of 8, standard deviation of 6
y_norm = 6 * np.random.randn(1000) + 8

# create an array of 1000 normally distributed values, mean of 5, standard deviation of 2
z_norm = 2 * np.random.randn(1000) + 5

pyplot.boxplot([w_norm,x_norm,y_norm,z_norm],labels=['w', 'x', 'y', 'z'])
pyplot.title('box plots')
pyplot.ylabel('values')
pyplot.show()
```

The red line in the box plots represents the median value of the data. The horizontal orange lines represent the 25th and 75th percentiles. This means that 50% of the data lives within the solid black box. The black lines extend out to the 0.35th and 99.65th percentile, so 99.3% of the data will be between the two black horizontal "whiskers". The open cirecles show any values that fall above the 99.65th percentile or below the 0.35th percentile.  

The box plots allow us to make a quick comparison of the 4 datasets. x has the largest median value, y has the largest spread, w has the smallest spread and smallest median value. All 4 distributions look roughly symmetric.

>Note: The minimum value of the data within each bin is usually plotted on the horizontal axis of histograms. The vertical axis can either be the number of data points that fall within a given bin (that's the default) or the density of a particular bin value. The total area of all the bars in a density histograms (bin width * bar height) will equal 1. This is often a good option to use when plotting more than one distribution in a single plot window.  By default the values of a distribution are plotted on the vertical axis of box plots. You can rotate the plot 90 degrees with the `vert` parameter (set to False).

Non-normal Distributions
---
One characteristic of the normal distribution is that it is symmetric. This is not true of all distributions. Below we've created a log-normal distribution and plotted its histogram. This distribution is very assymetric with most values clustered just above zero, and a few much larger values.

```python
# create an array of 1000 log-normally distributed values
x_lognorm = np.exp(1 * np.random.randn(1000) + 3)
pyplot.hist(x_lognorm, bins=20)
pyplot.title('log normal distribution example')
pyplot.xlabel('values')
pyplot.show()
```

>Note: As a general rule **always label your plots**. Sometimes this isn't absolutely necessary, for instance, if you're just doing quick exploration and will immiediately delete the plot and never look at it again you don't have to write the 2 or three extra lines to add a title and axis labels. However, if you plan to look at a plot again in the future, and especially if you plan to share your results with someone else **add a title and axes labels**.
